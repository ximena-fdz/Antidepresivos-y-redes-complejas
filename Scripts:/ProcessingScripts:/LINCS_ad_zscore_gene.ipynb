```python
import pandas as pd
import numpy as np
```

```python
# Cargar el archivo de niveles de expresión génica por instancia en Python
df = pd.read_csv("matriz_ad_GSE92742.txt", sep="\t", header=None, low_memory=False)

display(df)
```

```python
# Eliminar la primera columna del df
df = df.drop(df.columns[0], axis=1)
```

```python
#Resetear los índices de las columnas del df
df.columns = range(len(df.columns))

display(df)
```

```python
# Cargar el archivo de perturbaciones en Python

inst = pd.read_csv("GSE92742_Broad_LINCS_inst_info.txt", sep="\t", low_memory = False)

display(inst)
```

```python
# Mapear los datos de perturbaciones al df de niveles de expresión génica

# Definir filas con datos de 'inst'
nr_pert_id = inst['pert_id']
nr_pert_type = inst['pert_type']
nr_pert_dose = inst['pert_dose']
nr_pert_time = inst['pert_time']

# Agregar las nuevas filas a 'df'
df.loc[-1] = nr_pert_id
df.loc[-2] = nr_pert_type
df.loc[-3] = nr_pert_dose
df.loc[-4] = nr_pert_time

# Reordenar el índice para que sea continuo
df.index = range(len(df))

display(df)
```

```python
# Filtrar el df para obtener las instancias experimentales
filtered_exp = df.loc[:, df.loc[10906] == 'trt_cp']

display(filtered_exp)
```

```python
# Mantener sólo las instancias con dosis y tiempo de exposición máximos por perturbógeno experimental

# Hallar los valores únicos en la fila de 'pert_id'
unique_values = filtered_exp.loc[10905].unique()

# Inicializar una lista para almacenar las columnas que cumplen con la condición de valor máximo de dosis y tiempo de exposición
max_columns = []

# Iterar sobre cada valor único de 'pert_id'
for value in unique_values:
    # Encontrar las columnas asociadas a cada 'pert_id'
    columns = filtered_exp.columns[filtered_exp.loc[10905] == value]
    
    # Encontrar el valor máximo en las filas de 'pert_dose' y 'pert_time' para las columnas seleccionadas
    max_value_10907 = filtered_exp.loc[10907, columns].max()
    max_value_10908 = filtered_exp.loc[10908, columns].max()
    
    # Conservar las columnas donde los valores en las filas 'pert_dose' y 'pert_time' son máximos
    max_columns_10907 = columns[filtered_exp.loc[10907, columns] == max_value_10907]
    max_columns_10908 = columns[filtered_exp.loc[10908, columns] == max_value_10908]
    
    # Conservar sólo las columnas que cumplen ambas condiciones
    max_columns.extend(max_columns_10907.intersection(max_columns_10908))

# Crear el nuevo dataframe 'max_df' con las columnas seleccionadas
max_df = filtered_exp[max_columns]

display(max_df)
```

```python
# Filtrar el df para obtener las instancias control
filtered_ctl = df.loc[:, df.loc[10906] == 'ctl_vehicle']

display(filtered_ctl)
```

```python
# Seleccionar las filas asociadas a valores de expresión génica
filtered_ctl_subset = filtered_ctl.loc[1:10904, :]

# Convertir los valores a float64
filtered_ctl_subset = filtered_ctl_subset.apply(pd.to_numeric, errors='coerce')

display(filtered_ctl_subset)
```

```python
# Calcular la media y la desviación estándar por gen
mean_ctl = filtered_ctl_subset.mean(axis=1)
std_ctl = filtered_ctl_subset.std(axis=1)

# Añadir los valores a 'max_df'
max_df = max_df.assign(mean_ctl=mean_ctl, std_ctl=std_ctl)

display(max_df)
```

```python
# Copiar el dataframe y calcular los z-scores
zscore_exp = max_df.copy()

# Seleccionar las filas asociadas a valores de expresión génica
zscore_exp_subset = zscore_exp.iloc[1:10905, :]

# Convertir los valores a float64
zscore_exp_subset = zscore_exp_subset.apply(pd.to_numeric, errors='coerce')

display(zscore_exp_subset)
```

```python
# Calcular los z-scores de los niveles de expresión por gen

# Extraer las columnas de media y desviación estándar
mean_ctl = zscore_exp_subset['mean_ctl']
std_ctl = zscore_exp_subset['std_ctl']

# Eliminar las columnas de media y desviación estándar del subset
zscore_exp_subset = zscore_exp_subset.drop(columns=['mean_ctl', 'std_ctl'])

# Calcular el z-score de nivel de expresión por gen
zscores = (zscore_exp_subset.sub(mean_ctl, axis=0)).div(std_ctl, axis=0)

display(zscores)
```

```python
# Asignar los valores de 'zscores' a las filas correspondientes de 'zscore_exp'
zscore_exp.iloc[1:10905, :15018] = zscores.values

# Eliminar las columnas de media y desviación estándar
zscore_exp = zscore_exp.drop(columns=['mean_ctl', 'std_ctl'])

# Restablecer los índices de las columnas
zscore_exp.columns = range(zscore_exp.shape[1])

# Guardar el dataframe como un archivo .txt
zscore_exp.to_csv('zscore_exp.txt', sep='\t', index=False)

display(zscore_exp)
```

```python
# Extraer los valores únicos de 'pert_id'
unique_values = zscore_exp.iloc[10905].unique()

# Crear un diccionario para almacenar las medianas
medians_dict = {}

# Iterar sobre cada valor único de 'pert_id'
for value in unique_values:
    # Extraer las columnas asociadas al perturbógeno
    associated_columns = zscore_exp.columns[zscore_exp.iloc[10905] == value]
    
    # Tomar las filas correspondientes al z-score de esas columnas asociadas
    subset = zscore_exp.loc[1:10904, associated_columns]
    
    # Transformar los valores dtype de object a float64
    subset = subset.astype(float)
    
    # Calcular la mediana del z-score
    median_values = subset.median(axis=1)
    
    # Añadir la columna de medianas al diccionario
    medians_dict[value] = median_values
```

```python
# Convertir el diccionario a un dataframe
zscore_medians = pd.DataFrame(medians_dict)

# Guardar el dataframe como un archivo .txt
zscore_medians.to_csv('zscore_medians.txt', sep='\t', index=False)

display(zscore_medians)
```
